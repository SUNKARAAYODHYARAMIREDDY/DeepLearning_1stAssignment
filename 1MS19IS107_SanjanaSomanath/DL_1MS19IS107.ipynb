```python
data = data[data["c24"] == "SYNC"][data["c23"] == "ON"][data["c23"] == "ON"][data["c22"] == "ON"][data["c28"] == "ON"]
```

```python
data.drop(['c1'] + ['c' + str(n) for n in range(20,29)], axis=1, inplace = True)
data.head()
```

```python
data.describe()
```

```python
from sklearn import preprocessing

for i in range(2,18):
  plt.subplot(4,5,i - 1)
  plt.hist(data["c" + str(i)])
plt.show()
```

```python
data.median(axis = 0)
```

```python
plt.rcParams['figure.figsize'] = [50, 50]
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
newDf = min_max_scaler.fit_transform(data)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'b', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```

```python
from sklearn import preprocessing
z_scaler = preprocessing.MaxAbsScaler()
newDf = z_scaler.fit_transform(data)

newDf = pd.DataFrame(newDf)

for i in range(18):
  m = np.mean(newDf[i])
  cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'b', 'r')
  plt.subplot(4,5,i + 1)
  plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```
